```python
# 2. Lists
books = ["Python Basics", "Advanced Python", "Data Science with Python"]#Create list

# TODO: Add two more books to the list
books.append('Mysql Database')#add single element at the end of the list
books.insert(4,'Data Visualization with Power BI')#add single element at the end of the list

print("Updated book list:", books)#display result
```

```python
# 4. User Input
new_book = input("Enter a new book title: ")#take user input as book title
books.append(new_book)#add new book at the end of the list
print("Updated book list:", books)#display result
```

```python
# 5. Functions and Conditional Statements
def check_availability(book_name):
    # TODO: Check if the book is in the list and return True if available, False otherwise
    if book_name in list(books):
         return True
    else:
         return False
    pass

# Test the function
print(f"Is 'Python Basics' available? {check_availability('Python Basics')}")#display result
```

```python
# 6. File I/O
def save_books_to_file(filename):
    with open(filename, 'w') as f:#open file in write mode
        #Take each book from book list
        for book in books:
            f.write(f"{book}\n")#write book name into file

# TODO: Implement a function to read books from a file
def read_books_from_file(filename):
  #Handling file related exception using try-except
  try:
    with open(filename, 'r') as f:#open file in read mode
        return f.read().splitlines()#read file data
  except FileNotFoundError:
        print("File not found.")#display error message
        return None
  except Exception as e:
        print("Error loading file:", e)#display error message
        return None
  pass

# Test file I/O
save_books_to_file("books.txt")#call function
!cat books.txt  # Display file contents in Colab
```

```python
# 7. Pickle for serialization
def save_books_pickle(filename):
    with open(filename, 'wb') as f:#open file in write binary mode
        pickle.dump(books, f)#write books name into file

# TODO: Implement a function to load books using pickle
def load_books_pickle(filename):
    #Handling file related exception using try-except
    try:
        with open(filename, 'rb') as file:#open file in read binary mode
            books = pickle.load(file)#load data of file and assign to books
        return books # return books data
    except FileNotFoundError:
        print("File not found.")#display error merssage
        return None
    except Exception as e:
        print("Error loading file:", e)#display error message
        return None

# Test pickle functions
save_books_pickle("books.pkl")#call function
# Uncomment the next line after implementing load_books_pickle
print("Books loaded from pickle:", load_books_pickle("books.pkl"))#call function
```

```python
# 8. SQLite Database Interaction
def create_books_table():
    conn = sqlite3.connect('library.db')#connect database
    cursor = conn.cursor()#create cursor object
    cursor.execute('''CREATE TABLE IF NOT EXISTS books
        (id INTEGER PRIMARY KEY, title TEXT, quantity INTEGER)''')#execute query
    conn.commit()#commit query
    conn.close()#close database connection

# TODO: Implement a function to add a book to the SQLite database
def add_book_to_db(title, quantity):
    #Handling database connection and insert data related exceptions
    try:
        conn = sqlite3.connect('library.db')#create db connection
        cursor = conn.cursor()#create cursor object of database connection

        cursor.execute(
            '''INSERT INTO books (title, quantity) VALUES (?, ?)''',(title, quantity)
        )#write and execute insert query to insert new data

        conn.commit()#commit insert data
        print("Book added successfully!")

    except sqlite3.Error as e:
        print("Database error:", e)#display error message

    except Exception as e:
        print("Error:", e)#display error message

    finally:
        if conn:
            conn.close()#close db connection

# Test database functions
create_books_table()
# Uncomment the next line after implementing add_book_to_db
add_book_to_db("Python for Beginners", 10)
```

```python
# 9. Web API Interaction (GitHub API)
def get_python_repos():
    url = "https://api.github.com/search/repositories"#url path
    params = {"q": "language:python", "sort": "stars", "order": "desc"}#url parameters
    response = requests.get(url, params=params)#get url data
    data = response.json()#take read data in json form
    return [repo['name'] for repo in data['items'][:5]]#display all read data

# Test API function
print("Top 5 Python repositories on GitHub:", get_python_repos())#call function

print("\nLibrary Management System operations completed.")
```
