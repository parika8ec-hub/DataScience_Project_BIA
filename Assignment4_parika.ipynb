```python
# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:
# Concatenate df1 and df2 vertically
df_vertical = pd.concat([df1, df2], axis=0, ignore_index=True)

print("Vertical Concatenation:")
print(df_vertical)#display result

# Merge df1 and df3 horizontally on the 'Key' column
df_merged = pd.merge(df1, df3, left_index=True, right_index=True)

print("Merge Horizontally:")
print(df_merged)#display result
```

```python
# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:
import math

# Create function to calculate square
def square_number(x):
    return math.pow(x, 2)  # math.pow returns float

# Call the function
print("Square using function of 5:", square_number(5))

# Lambda function to calculate square
square_lambda = lambda x: math.pow(x, 2)

# Call the lambda function
print("Square using lambda of 5:", square_lambda(5))
```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# List of first 5 even numbers
even_numbers = [2, 4, 6, 8, 10]

# Create an iterator
even_iter = iter(even_numbers)

print("Iterator output:")
#Take each element of iterator
for num in even_iter:
    print(num)

# Generator function for first n even numbers
def even_generator(n):
    for i in range(1, n+1):
        yield 2 * i

# Create generator by calling even_generator() function
gen = even_generator(5)

print("\nGenerator output:")
for num in gen:
    print(num)
```

```python
# Task 10: Map, Reduce and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
#Use map to square all numbers in a list
squared_numbers = list(map(lambda x: x**2, numbers))
print("Squared numbers using map:", squared_numbers)#display result

#Use reduce to find the product of all numbers in a list
product = reduce(lambda x, y: x * y, numbers)
print("Product of all numbers using reduce:", product)#display result

#Use filter to get only even numbers from a list
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers using filter:", even_numbers)#display result
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
# Define the Rectangle class
class Rectangle:
    #Create constructor of class
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Method to calculate area
    def area(self):
        return self.length * self.width

    # Method to calculate perimeter
    def perimeter(self):
        return 2 * (self.length + self.width)

# Create an object of Rectangle
obj1 = Rectangle(5, 3)

# Access and display data member of class
print("Length of rectangle:", obj1.length)
print("Width of rectangle:", obj1.width)

# Call methods and display result
print("\nArea of rectangle:", obj1.area())
print("Perimeter of rectangle:", obj1.perimeter())
```

```python
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
#Find average salary by department
avg_salary_by_dept = df_employees.groupby('Department')['Salary'].mean()
print("Average salary by department:")
print(avg_salary_by_dept)

#Get names of employees with salary > 60000
high_salary_employees = df_employees[df_employees['Salary'] > 60000]['Name'].tolist()
print("\nEmployees with salary > 60000:")
print(high_salary_employees)

#Add a new column 'Bonus' which is 10% of salary
df_employees['Bonus'] = df_employees['Salary'] * 0.10
print("\nDataFrame with Bonus column:")
print(df_employees)
```
